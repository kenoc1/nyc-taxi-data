```python
from datetime import *
import pandas as pd
import os

for filename in os.listdir("../data/taxi_data/"):
    if filename.endswith(".csv") and filename.startswith("yellow_tripdata_2021"):
        df = pd.read_csv(f'../data/taxi_data/{filename}', low_memory=False)
        df.tpep_pickup_datetime = df.tpep_pickup_datetime.apply(
            lambda x: pd.to_datetime(x, format='%Y-%m-%d %H:%M:%S').strftime('%d/%m/%Y'))
        df2 = df.groupby(['tpep_pickup_datetime'])['tpep_pickup_datetime'].count()
        df2.to_csv(f'../data/taxi_data/{filename}_customized.csv', sep=',')
        print(f'{filename} transformed')




# print(pd.to_datetime("2020-04-01 00:41:22", format='%Y-%m-%d %H:%M:%S').strftime('%Y-%m-%d'))
#
# df.tpep_pickup_datetime = df.tpep_pickup_datetime.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d %H:%M:%S').strftime('%d/%m/%Y'))
# df2 = df.groupby(['tpep_pickup_datetime'])['tpep_pickup_datetime'].count()
# df2.to_csv('../data/taxi_data/yellow_tripdata_2020-10_customized.csv', sep=',')
# df2
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

df_corona_cases = pd.read_csv('../data/coronavirus-data-master/trends/tests.csv')

df_whole_year_2021 = []
for filename in os.listdir("../data/taxi_data/"):
    if filename.endswith("customized.csv") and filename.startswith("yellow_tripdata_2021"):
        df_single_month = pd.read_csv(f'../data/taxi_data/{filename}', low_memory=False)
        df_whole_year_2021.append(df_single_month)

df_whole_year_2021 = pd.concat(df_whole_year_2021)

df_full = pd.merge(df_corona_cases, df_whole_year_2021,
                   on='DATE',
                   how='inner')
df_full = df_full.drop('INCOMPLETE', axis=1)

#Generate correlation map
corrMatrix = df_full.corr()
plt.figure(figsize=(16, 5))
sn.heatmap(corrMatrix, annot=True)
plt.show()




```

```python
#sn.lineplot(df_full['DATE'], df_full['TOTAL_TESTS'])
sn.set(rc={'figure.figsize': (20, 20)})
ax = sn.lineplot(df_full['DATE'], y=df_full['TOTAL_TESTS'], data=df_full, legend='brief', label='TOTAL_TESTS')
ax2 = sn.lineplot(df_full['DATE'], y=df_full['Number_OF_RIDES'], data=df_full, legend='brief', label='Number_OF_RIDES')
ax3 = sn.lineplot(df_full['DATE'], y=df_full['POSITIVE_TESTS'], data=df_full, legend='brief', label='POSITIVE_TESTS')
plt.show()
```
