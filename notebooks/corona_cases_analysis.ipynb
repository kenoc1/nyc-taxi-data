```python
from datetime import *
import pandas as pd

df = pd.read_csv("../data/taxi_data/yellow_tripdata_2020-04.csv", low_memory=False)
#print(pd.to_datetime("2020-04-01 00:41:22", format='%Y-%m-%d %H:%M:%S').strftime('%Y-%m-%d'))

df.tpep_pickup_datetime = df.tpep_pickup_datetime.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d %H:%M:%S').strftime('%d/%m/%Y'))
df2 = df.groupby(['tpep_pickup_datetime'])['tpep_pickup_datetime'].count()
df2.to_csv('../data/taxi_data/yellow_tripdata_2020-04_customized.csv', sep=',')
df2
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
df = pd.read_csv("../data/taxi_data/yellow_tripdata_2020-04_customized.csv", low_memory=False)
df2 = pd.read_csv('../data/coronavirus-data-master/trends/tests.csv')

df_full = pd.merge(df, df2,
                   on='DATE',
                   how='inner')


# Generate correlation map
corrMatrix = df_full.corr()
plt.figure(figsize = (16,5))
sn.heatmap(corrMatrix, annot=True)
plt.show()
```
